```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('listings.csv', sep=',',header='infer')
df.values
```

```python
df.head(10)
```

```python
df.count()
```

```python
hoods = df['neighbourhood'].value_counts()\
                 .to_frame('count').rename_axis('neighbourhoods')\
                 .reset_index()
        
hoods
```

```python
hoods.plot(kind='bar',x='neighbourhoods', y='count')
plt.show()
```
